Load Data

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Display plots inline
%matplotlib inline

# Load data
df = pd.read_csv("../data/raw/data.csv")
df.head()
```

Data Overview

```python
print("Dataset Shape:", df.shape)
print("\nColumn Data Types:")
print(df.dtypes)
```

```python
#Summary Statistics
print("\nSummary Statistics for Numerical Columns:")
display(df.describe())
```

Missing Values

```python
missing_values = df.isnull().sum()
print("\nMissing Values per Column:")
print(missing_values)

# Missing value handling strategy (if any missing)
# In this dataset, there are no missing values.
# If missing, we would use:
# - Median for numerical columns with outliers
# - Mode for categorical columns
# - Drop columns with >30% missing data
```

Numerical Distributions

```python
numerical_cols = ['Amount', 'Value', 'PricingStrategy', 'FraudResult']
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], bins=50, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

Outlier Detection (Numerical Features)

```python
# Outlier Detection (Numerical Features)
for col in ['Amount', 'Value']:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

Categorical Feature Analysis

```python
categorical_cols = ['CurrencyCode', 'ProductCategory', 'ChannelId', 'ProviderId']
for col in categorical_cols:
    plt.figure(figsize=(10,4))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Frequency of {col}')
    plt.show()
```

Correlation Analysis (Numerical)

```python
plt.figure(figsize=(8,6))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Numerical Features")
plt.show()
```

Top 5 Insights

```python
#Top 5 Insights
print("\nKey Insights from EDA:")
insights = [
    "1. The 'Amount' and 'Value' columns show heavy skew with extreme outliers, "
    "suggesting the need for log transformation or robust scaling before modeling.",
    "2. Most transactions are non-fraudulent (FraudResult=0), indicating a highly imbalanced target if used.",
    "3. The 'ProductCategory' is dominated by a few categories such as airtime and financial services, "
    "which could influence feature importance in a credit risk model.",
    "4. 'Amount' and 'Value' are highly correlated (~1.0), so one can be used to reduce multicollinearity.",
    "5. Missing values are absent, simplifying preprocessing, but careful handling of outliers is necessary."
]

for insight in insights:
    print(insight)
```
